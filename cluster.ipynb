```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('house_data.csv')
df
```

```python
df.shape
```

```python
df.drop(['sr no','Sqft','Floor','TotalFloor'], axis=1)
```

```python
#we are forming cluster on the basis of house specifications thats why we have droped the 'Sqft','Floor','TotalFloor' coloumns 
from sklearn.cluster import KMeans
km=KMeans(n_clusters=6)
yp=km.fit_predict(df)
```

```python
df['house_feature_cluster']=yp
df.head(2)
```

```python
df1=df[df.house_feature_cluster==0]
df2=df[df.house_feature_cluster==1]
df3=df[df.house_feature_cluster==2]
df4=df[df.house_feature_cluster==3]
df5=df[df.house_feature_cluster==4]
df6=df[df.house_feature_cluster==5]
```

```python
plt.scatter(df1['Bedroom'],df1['Price'],color='black')
plt.scatter(df2['Bedroom'],df2['Price'],color='violet')
plt.scatter(df3['Bedroom'],df3['Price'],color='blue')
plt.scatter(df4['Bedroom'],df4['Price'],color='green')
plt.scatter(df5['Bedroom'],df5['Price'],color='yellow')
plt.scatter(df6['Bedroom'],df6['Price'],color='red')
```

```python
wcss = []

# Iterate over a range of cluster values
for k in range(1, 11):
    km = KMeans(n_clusters=k)
    km.fit(df)
    wcss.append(km.inertia_)  # Append the WCSS value to the list

# Plot the WCSS values
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```
